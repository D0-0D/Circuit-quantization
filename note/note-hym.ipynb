```python
import sympy as sp
import networkx as nx

class Component:
    def __init__(self, u:int, v:int, tp:str, value):
        self.u = u
        self.v = v
        self.type = tp.upper()
        valid_types = {'C', 'L', 'JJ'}
        if self.type not in valid_types:
            raise ValueError(f"元件类型 '{self.type}' 不在支持列表 {valid_types} 中")
        if isinstance(value, str):
            self.value = sp.symbols(value)
        else:
            self.value = value
    @property
    def weight(self):
        #连支中尽量多C就是生成树尽量少C, 所以C的wt应该是1, 其它的是2
        return 1 if self.type == 'C' else 2
    

def build_circuit_graph(components:list[Component]):
    """
    根据元件列表构建电路图。
    参数 components 格式: [(node1, node2, type, value), ...]
    返回: G (MultiGraph), edge_map (dict)
    """
    G = nx.MultiGraph()
    edge_map = {}

    # 遍历列表并添加边
    for i, comp in enumerate(components):
        G.add_edge(
            comp.u, 
            comp.v, 
            key=i, 
            type=comp.type, 
            value=comp.value, 
            name=comp.name,
            weight=comp.weight
        )       
        edge_map[i] = (comp.u, comp.v, i)
    T = nx.minimum_spanning_tree(G, weight='weight')

    return T,G, edge_map
```

```python



my_components = [
    (0, 1, 'JJ', 0.5),   
    (1, 2, 'L', 10e-3),  
    (2, 0, 'C', 0.5),   
    (1, 0, 'C', 2e-6),   
]

T,G, edge_map = build_circuit_graph(my_components)


print(f"生成树支路 ID: {[key for u, v, key in T.edges(keys=True)]}")

T.edges
```
